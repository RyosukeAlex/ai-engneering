```python
# ==== ライブラリ =====
import pandas as pd
import numpy as np
```

```python
# ===== ANDゲート ======
def AND(x1, x2):
    # y = b + (x1 * w1) + (x2 * w2)をパーセプトロンの動作式として、0を超えたら発火させる。
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = - 0.7
    y = b + np.sum(x * w)
    if y <= 0:
        return 0
    else :
        return 1

print(AND(23, 0.1))

# 3つのゲートはどれも「まったく同じ計算式」で動いていて、変わるのは数字(重みとバイアス)だけ
```
