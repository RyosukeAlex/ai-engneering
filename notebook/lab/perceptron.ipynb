```python
# ==== ライブラリ =====
import pandas as pd
import numpy as np
```

```python
# ===== ANDゲート ======
def AND(x1, x2):
    # y = b + (x1 * w1) + (x2 * w2)をパーセプトロンの動作式として、0を超えたら発火させる。
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = - 0.7
    y = b + np.sum(x * w)
    if y <= 0:
        return 0
    else :
        return 1

print(AND(23, 0.1))

# 3つのゲートはどれも「まったく同じ計算式」で動いていて、変わるのは数字(重みとバイアス)だけ
```

```python
 # ===== NANDゲート =====
def NAND(x1, x2):
    # ANDゲートと動作式は同様で、y = b + (x1 * w1) + (x2 * w2)を利用する。
    # ANDゲートとの相違点は、重みとバイアスのみ。
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    y = b + np.sum(x * w)
    if y <= 0:
        return 0
    else:
        return 1

print(NAND(43,39))
```

```python
 # ===== ORゲート =====
def OR(x1, x2):
    # ANDゲートと動作式は同様で、y = b + (x1 * w1) + (x2 * w2)を利用する。
    # ANDゲートとの相違点は、重みとバイアスのみ。
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    y = b + np.sum(x * w)
    if y <= 0:
        return 0
    else:
        return 1

print(NAND(43,39))
```

```python
# ===== XORゲート =====
def XOR(x1, x2):
    # AND, NAND, ORゲートを使って、XORゲートを実装することができる。
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(3, 4))
```
